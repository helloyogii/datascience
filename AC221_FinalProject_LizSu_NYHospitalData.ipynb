# 1. Predictive Models

```python
#importing packages required for models
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
import statsmodels.api as sm
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
```

```python
#splitting data into independent and dependent variables
yTrain = df.Length_of_Stay
xTrain = df.drop('Length_of_Stay', axis=1)
xTrain.head()
yTrain.head()
xTrain.shape
yTrain.shape
```

```python
#converting independent variables to string to use Label Encoder
xTrain = xTrain.astype('str')
```

```python
#using Label Encoder to convert categorical variables into regression format
le = preprocessing.LabelEncoder()
for i in range(36):
    xTrain.iloc[:,i] = le.fit_transform(xTrain.iloc[:,i])
```

```python
xTrain.head()
yTrain.head()
```

```python
#splitting the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(xTrain, yTrain, train_size=0.8, test_size=0.2, random_state=0)
```

```python
#fitting and summarizing the linear regression model
import statsmodels.api as sm
linear_regression = sm.OLS(y_train, x_train)
model = linear_regression.fit()
y_pred_lr = model.predict(x_test)
model.summary()
```

```python
#defining a model performance function
def pred_summary(y_test,y_pred):
    print("R squared is:",r2_score(y_test, y_pred))
    print("MAE error is:",mean_absolute_error(y_test, y_pred))
```

```python
#evaluating performance
pred_summary(y_test, y_pred_lr)
```

## Lasso Model

```python
#fitting a Lasso model
alphas = [0.001,0.01,0.1,1,10]

lasso_model = LassoCV(alphas = alphas, max_iter = 100000, fit_intercept = True, cv = 5)
lasso_model.fit(x_train,y_train)
print(lasso_model.alpha_)
y_pred_lasso = lasso_model.predict(x_test)

#evaluating performance
pred_summary(y_test, y_pred_lasso)
```

# Ridge Model 

```python
#fitting a Ridge model
ridge_model = RidgeCV(alphas = alphas,fit_intercept = True,cv = 5)
ridge_model.fit(x_train,y_train)
print(ridge_model.alpha_)
y_pred_ridge = ridge_model.predict(x_test)

#evaluating performance
pred_summary(y_test, y_pred_ridge)
```

# Random Forest

```python
#fitting a Random Forest model
rf_model = RandomForestRegressor(n_estimators=10, n_jobs=-1).fit(x_train,y_train)
y_pred_rf = rf_model.predict(x_test)

#evaluating performance
pred_summary(y_test, y_pred_rf)
```

```python
#determining the top 15 most important features
feat_importances = pd.Series(rf_model.feature_importances_, index=x_train.columns)
feat_importances.nlargest(15).plot(kind='barh')
```

```python
#plotting MAE across models
error = pd.DataFrame(data = {"Linear Regression":mean_absolute_error(y_test, y_pred_lr),
                               "LASSO":mean_absolute_error(y_test, y_pred_lasso),
                               "Ridge":mean_absolute_error(y_test, y_pred_ridge),
                               "Random Forest":mean_absolute_error(y_test, y_pred_rf)},index = ['MAE'])
error.plot(kind='barh')
```

# 2. Exploration of k-anonymity

```python
#creating a qi_key column 
df['qi_key'] = df.astype(str).apply(lambda row: '_'.join(row.values), axis=1)
```

```python
df.head()
```

```python
qi_key_counts = df.qi_key.value_counts().rename_axis('qi_key').reset_index(name='counts')
qi_key_counts
```

```python
#there are 2.3 million qi_keys that only have 1 count! this dataset does not satisfy k-anonymity.
qi_key_counts.query('counts < 2')
```

```python
#per professor waldo's feedback from the presentation, looking at the portion of the dataset that represents abortions
df3 = df[df["Health_Service_Area"].isnull()]
df3.head()
```

```python
df3.is_copy = None
df3['qi_key'] = df3.astype(str).apply(lambda row: '_'.join(row.values), axis=1)
```

```python
#the subset of abortions does not do a good job at k-anonymity either, with all records being unique
abortion_qi_key_counts = df3.qi_key.value_counts().rename_axis('qi_key').reset_index(name='counts')
abortion_qi_key_counts
```

From the above exploration of privacy, this dataset is not well-protected from adversarial actions from a k-anonymity or l-diversity perspective. Malicious actors, if they were able to get their hands on another dataset, could very well engineer the identities of some of these records.
